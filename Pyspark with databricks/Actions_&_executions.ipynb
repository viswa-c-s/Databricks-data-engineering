```python
df = spark.read.table("dev_data.test.Products")
```

```python
display(df)
```

```python
# Transformations -> builds logical plan -> Lazy(nothing runs) evaluation
#
# examples : select() , filter(), withColumn(), groupBy(), orderBy(), join(), union(), intersect(), except(), dropDuplicates(), sortWithinPartitions(), repartition(), coalesce(), limit(),  toDF(), withWatermark(), 
# 

# Lazy -> execution/storage hints (do not trigger execution) cache(), checkpoint(), persist(), unpersist(),
```

```python
# Actions are trigger execution
# Returns result to driver
# Examples: count(), first(), take(), show(), collect(), save(), saveAsTable(), saveAsParquetFile(), saveAsCsvFile(), saveAsJsonFile(), write() 

# Spark executes tranformations only when an action is called/triggered.
```

```python
# count() is an action and it's safe to use
# it returns number of rows in the dataframe, safe use case for larger datasets
# used to validate the number of rows in the dataframe, data quality check and logging
df.count()
```

```python
# collect()
# returns all the rows in the dataframe as a list of Row objects
# brings all data to driver so it can cause OOM
# use case: small datasets, data quality check
# collect should be avoided on large datasets as it loads all data into driver
df.collect()
```

```python
# take() is like show returns first n rows, good for debugging
# it returns a list of Row objects
df.take(3)
```

```python
# explain()

from pyspark.sql.functions import col

df.filter(col("Price") > 40000).explain()
```
